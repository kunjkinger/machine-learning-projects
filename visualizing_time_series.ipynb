```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/starbucks.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df.index
```

```python
df.plot()
```

```python
title = 'title'
ylabel = 'ylabel'
xlabel = 'xlabel'
```

```python
ax = df['Close'].plot(figsize=(12,5),title=title)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel,ylabel=ylabel)
```

```python
#option1
df['Close']['2017-01-01':'2017-12-31'].plot(figsize=(12,4))
```

```python
#option 2
df['Close'].plot(figsize=(12,4),xlim=['2017-01-01','2017-12-31'],ylim=[50,62])
```
